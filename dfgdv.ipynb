```python
# Подключаем библиотеки
import parselmouth
import matplotlib.pyplot as plt
from IPython.display import Audio, display, HTML, Javascript
from google.colab import output, files
import numpy as np
import base64
import os

print("="*70)
print("🎤 АНАЛИЗАТОР ИНТОНАЦИИ (метод: Filtered Autocorrelation, Гц (log))")
print("="*70)

# Выбор режима работы
mode_html = '''
<div style="margin: 20px 0; padding: 20px; background-color: #f0f0f0; border-radius: 10px;">
    <h3>📌 Выберите способ загрузки аудио:</h3>
    <button id="record-mode-btn" style="
        background-color: #ff4444;
        color: white;
        font-size: 18px;
        padding: 15px 30px;
        border: none;
        border-radius: 50px;
        cursor: pointer;
        margin: 10px;
        box-shadow: 0 4px 8px rgba(0,0,0,0.2);
    ">
        🎤 Записать с микрофона
    </button>

    <button id="upload-mode-btn" style="
        background-color: #4CAF50;
        color: white;
        font-size: 18px;
        padding: 15px 30px;
        border: none;
        border-radius: 50px;
        cursor: pointer;
        margin: 10px;
        box-shadow: 0 4px 8px rgba(0,0,0,0.2);
    ">
        📁 Загрузить файл (MP3, MP4, WAV)
    </button>

    <p id="mode-status" style="font-size: 16px; margin-top: 15px;">⚪ Выберите режим</p>
</div>

<!-- Интерфейс для настройки параметров -->
<div style="margin: 20px 0; padding: 20px; background-color: #e0e0ff; border-radius: 10px;">
    <h3>⚙️ Настройки анализа (как в Praat):</h3>
    <label>Pitch floor (мин. частота): </label>
    <input type="number" id="pitch-floor" value="75" min="40" max="200" step="5"> Гц<br>
    <label>Pitch ceiling (макс. частота): </label>
    <input type="number" id="pitch-ceiling" value="600" min="200" max="800" step="10"> Гц<br>
    <p style="font-size: 14px; color: #666;">
        💡 Для мужского голоса: floor=75, ceiling=300<br>
        💡 Для женского голоса: floor=100, ceiling=600
    </p>
</div>

<!-- Интерфейс для записи (изначально скрыт) -->
<div id="recorder-interface" style="display: none;">
    <div style="padding: 20px; background-color: #ffe6e6; border-radius: 10px;">
        <h3>🔴 Запись с микрофона</h3>
        <button id="record-btn" style="
            background-color: #ff4444;
            color: white;
            font-size: 20px;
            padding: 15px 30px;
            border: none;
            border-radius: 50px;
            cursor: pointer;
            margin: 10px;
        ">
            🔴 Начать запись
        </button>

        <button id="stop-btn" style="
            background-color: #666;
            color: white;
            font-size: 20px;
            padding: 15px 30px;
            border: none;
            border-radius: 50px;
            cursor: pointer;
            margin: 10px;
            display: none;
        ">
            ⏹️ Остановить запись
        </button>

        <p id="record-status" style="font-size: 18px;">⚪ Готов к записи</p>
    </div>
</div>

<!-- Интерфейс для загрузки (изначально скрыт) -->
<div id="upload-interface" style="display: none;">
    <div style="padding: 20px; background-color: #e6f7e6; border-radius: 10px;">
        <h3>📁 Загрузка файла</h3>
        <button id="upload-btn" style="
            background-color: #4CAF50;
            color: white;
            font-size: 20px;
            padding: 15px 30px;
            border: none;
            border-radius: 50px;
            cursor: pointer;
            margin: 10px;
        ">
            📂 Выбрать файл
        </button>
        <p id="upload-status" style="font-size: 18px;">⚪ Файл не выбран</p>
    </div>
</div>

<script>
// Переменные для записи
let mediaRecorder;
let audioChunks = [];
let stream;

// Получить значения настроек
function getPitchSettings() {
    return {
        floor: parseFloat(document.getElementById('pitch-floor').value),
        ceiling: parseFloat(document.getElementById('pitch-ceiling').value)
    };
}

// Показать интерфейс записи
document.getElementById('record-mode-btn').onclick = function() {
    document.getElementById('recorder-interface').style.display = 'block';
    document.getElementById('upload-interface').style.display = 'none';
    document.getElementById('mode-status').innerHTML = '✅ Выбран режим: запись с микрофона';
};

// Показать интерфейс загрузки
document.getElementById('upload-mode-btn').onclick = function() {
    document.getElementById('recorder-interface').style.display = 'none';
    document.getElementById('upload-interface').style.display = 'block';
    document.getElementById('mode-status').innerHTML = '✅ Выбран режим: загрузка файла';
};

// Логика записи
document.getElementById('record-btn').onclick = async function() {
    try {
        stream = await navigator.mediaDevices.getUserMedia({ audio: true });
        mediaRecorder = new MediaRecorder(stream);
        audioChunks = [];

        mediaRecorder.ondataavailable = event => {
            audioChunks.push(event.data);
        };

        mediaRecorder.onstop = async () => {
            const settings = getPitchSettings();
            const audioBlob = new Blob(audioChunks, { type: 'audio/wav' });
            const reader = new FileReader();
            reader.readAsDataURL(audioBlob);
            reader.onloadend = () => {
                const base64data = reader.result.split(',')[1];
                google.colab.kernel.invokeFunction('process_audio',
                                                  [base64data, 'recorded', settings.floor, settings.ceiling], {});
            };

            stream.getTracks().forEach(track => track.stop());

            document.getElementById('record-status').innerHTML = '✅ Запись завершена! Анализирую...';
            document.getElementById('record-btn').style.display = 'inline';
            document.getElementById('stop-btn').style.display = 'none';
        };

        mediaRecorder.start();

        document.getElementById('record-btn').style.display = 'none';
        document.getElementById('stop-btn').style.display = 'inline';
        document.getElementById('record-status').innerHTML = '🔴 Идет запись... Говорите!';

    } catch (err) {
        document.getElementById('record-status').innerHTML = '❌ Ошибка: ' + err.message;
    }
};

document.getElementById('stop-btn').onclick = function() {
    if (mediaRecorder && mediaRecorder.state === 'recording') {
        mediaRecorder.stop();
    }
};

// Логика загрузки файла
document.getElementById('upload-btn').onclick = function() {
    const settings = getPitchSettings();
    google.colab.kernel.invokeFunction('select_file', [settings.floor, settings.ceiling], {});
};
</script>
'''

# Функция для обработки загруженного файла
def process_uploaded(filename, pitch_floor, pitch_ceiling):
    try:
        print(f"\n📂 Обрабатываю файл: {filename}")
        print(f"⚙️ Настройки: floor={pitch_floor} Hz, ceiling={pitch_ceiling} Hz")

        # Определяем тип файла по расширению
        ext = os.path.splitext(filename)[1].lower()

        if ext == '.wav':
            # Прямая загрузка WAV
            sound = parselmouth.Sound(filename)
        else:
            # Конвертируем MP3/MP4 в WAV
            output_wav = 'temp_uploaded.wav'
            !ffmpeg -i "{filename}" -acodec pcm_s16le -ar 44100 "{output_wav}" -y 2>/dev/null
            sound = parselmouth.Sound(output_wav)
            filename = output_wav

        analyze_sound(sound, filename, pitch_floor, pitch_ceiling)

    except Exception as e:
        print(f"❌ Ошибка при обработке файла: {e}")
        import traceback
        traceback.print_exc()

# Функция для выбора файла
def select_file(pitch_floor, pitch_ceiling):
    from google.colab import files
    uploaded = files.upload()
    for filename in uploaded.keys():
        process_uploaded(filename, pitch_floor, pitch_ceiling)

# Функция для обработки записанного аудио
def process_recorded(audio_base64, source, pitch_floor, pitch_ceiling):
    try:
        # Декодируем аудио из base64
        audio_bytes = base64.b64decode(audio_base64)

        # Сохраняем временный файл
        with open('temp_recorded.webm', 'wb') as f:
            f.write(audio_bytes)

        # Конвертируем в WAV
        !ffmpeg -i temp_recorded.webm -acodec pcm_s16le -ar 44100 temp_recorded.wav -y 2>/dev/null

        # Загружаем звук
        sound = parselmouth.Sound('temp_recorded.wav')

        analyze_sound(sound, 'temp_recorded.wav', pitch_floor, pitch_ceiling)

    except Exception as e:
        print(f"❌ Ошибка при анализе записи: {e}")
        import traceback
        traceback.print_exc()

# ОСНОВНАЯ ФУНКЦИЯ АНАЛИЗА
def analyze_sound(sound, filename, pitch_floor, pitch_ceiling):
    print("\n" + "="*60)
    print(f"🔬 АНАЛИЗ ЗВУКА (метод: Filtered Autocorrelation)")
    print(f"📊 Диапазон: {pitch_floor}-{pitch_ceiling} Hz")
    print("="*60)

    # --- АНАЛИЗ ИНТОНАЦИИ С FILTERED AUTOCORRELATION ---
    pitch = sound.to_pitch(
        time_step=0.01,
        pitch_floor=pitch_floor,
        pitch_ceiling=pitch_ceiling
    )

    # Получаем все значения (включая паузы где частота = 0)
    pitch_values = pitch.selected_array['frequency']
    times_pitch = pitch.xs()

    # Для отображения с паузами используем все точки
    all_pitch = pitch_values.copy()
    all_times = times_pitch.copy()

    # Заменяем 0 на NaN для разрывов линии
    all_pitch[all_pitch == 0] = np.nan

    # Для статистики используем только голосовые участки
    valid_indices = pitch_values > 0
    clean_pitch = pitch_values[valid_indices]
    clean_times = times_pitch[valid_indices]

    if len(clean_pitch) == 0:
        print("❌ Голос не распознан. Попробуйте изменить настройки:")
        print(f"   Текущие: floor={pitch_floor} Hz, ceiling={pitch_ceiling} Hz")
        print("   Для мужского голоса попробуйте floor=60, ceiling=300")
        print("   Для женского голоса попробуйте floor=100, ceiling=600")
        return

    # Статистика интонации
    max_f0 = np.max(clean_pitch)
    min_f0 = np.min(clean_pitch)
    mean_f0 = np.mean(clean_pitch)
    median_f0 = np.median(clean_pitch)
    std_f0 = np.std(clean_pitch)
    max_time = clean_times[np.argmax(clean_pitch)]

    # --- СПЕКТРОГРАММА ---
    spectrogram = sound.to_spectrogram(window_length=0.03, time_step=0.002)
    spectrogram_values = spectrogram.values
    frequencies = spectrogram.y_grid()
    times_spec = spectrogram.x_grid()

    # Конвертируем в децибелы
    spectrogram_db = 10 * np.log10(spectrogram_values + 1e-10)

    # Ограничиваем частоту для лучшей визуализации
    freq_limit_idx = np.argmin(np.abs(frequencies - 5000))
    frequencies_limited = frequencies[:freq_limit_idx]
    spectrogram_limited = spectrogram_db[:freq_limit_idx, :]

    # --- СОЗДАЕМ ГРАФИКИ ---
    plt.figure(figsize=(20, 12))

    # 1. СПЕКТРОГРАММА + ИНТОНАЦИЯ
    plt.subplot(3, 2, 1)
    plt.imshow(spectrogram_limited, aspect='auto', origin='lower',
               extent=[times_spec[0], times_spec[-1],
                      frequencies_limited[0], frequencies_limited[-1]],
               cmap='magma', interpolation='bilinear')
    plt.colorbar(label='Интенсивность (дБ)')

    # Накладываем интонацию (с паузами)
    plt.semilogy(all_times, all_pitch, 'c-', linewidth=2.5, label='Интонация')
    if not np.isnan(max_f0):
        plt.semilogy(max_time, max_f0, 'wo', markersize=8, label=f'Пик: {max_f0:.1f} Гц')

    plt.xlabel('Время (секунды)')
    plt.ylabel('Частота (Гц, log шкала)')
    plt.title(f'🔬 СПЕКТРОГРАММА + ИНТОНАЦИЯ (Filtered Autocorrelation, {pitch_floor}-{pitch_ceiling} Hz)')
    plt.ylim(pitch_floor, min(5000, pitch_ceiling * 1.5))
    plt.grid(True, alpha=0.3, which='both')
    plt.legend(loc='upper right')

    # 2. ИНТОНАЦИОННЫЙ КОНТУР С ПАУЗАМИ
    plt.subplot(3, 2, 2)
    # Рисуем линию с разрывами на паузах
    plt.semilogy(all_times, all_pitch, 'b-', linewidth=2.5, label='Интонация')

    # Отмечаем паузы красными точками
    pause_indices = pitch_values == 0
    if np.any(pause_indices):
        plt.semilogy(times_pitch[pause_indices], [pitch_floor] * np.sum(pause_indices),
                    'rx', markersize=4, label=f'Паузы ({np.sum(pause_indices)})')

    if not np.isnan(max_f0):
        plt.semilogy(max_time, max_f0, 'ro', markersize=10, label=f'Макс: {max_f0:.1f} Гц')
    plt.axhline(y=mean_f0, color='g', linestyle='--', alpha=0.5, label=f'Среднее: {mean_f0:.1f} Гц')

    # Заполняем область где есть голос
    plt.fill_between(all_times, pitch_floor, all_pitch, where=~np.isnan(all_pitch),
                     color='blue', alpha=0.1)

    plt.xlabel('Время (секунды)')
    plt.ylabel('Частота (Гц, log шкала)')
    plt.title(f'📈 ИНТОНАЦИОННЫЙ КОНТУР С ПАУЗАМИ ({np.sum(pause_indices)} пауз)')
    plt.ylim(pitch_floor, pitch_ceiling)
    plt.grid(True, alpha=0.3, which='both')
    plt.legend(loc='upper right')

    # 3. ГИСТОГРАММА РАСПРЕДЕЛЕНИЯ ЧАСТОТ
    plt.subplot(3, 2, 3)
    plt.hist(clean_pitch, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.axvline(x=mean_f0, color='r', linestyle='--', linewidth=2, label=f'Среднее')
    plt.axvline(x=median_f0, color='g', linestyle='--', linewidth=2, label=f'Медиана')
    plt.xlabel('Частота (Гц)')
    plt.ylabel('Количество')
    plt.title('📊 РАСПРЕДЕЛЕНИЕ ЧАСТОТ')
    plt.legend()

    # 4. СТАТИСТИКА (столбцы)
    plt.subplot(3, 2, 4)
    bars = plt.bar(['Минимум', 'Среднее', 'Медиана', 'Максимум'],
                  [min_f0, mean_f0, median_f0, max_f0],
                  color=['green', 'blue', 'orange', 'red'])
    plt.ylabel('Частота (Гц)')
    plt.title('📊 ХАРАКТЕРИСТИКИ ГОЛОСА')

    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.1f} Гц', ha='center', va='bottom')

    # 5. ФОРМА ВОЛНЫ
    plt.subplot(3, 2, 5)
    time_wave = np.linspace(0, len(sound.values[0])/sound.sampling_frequency, len(sound.values[0]))
    plt.plot(time_wave, sound.values[0], 'b-', linewidth=0.5)

    # Отмечаем паузы на волне
    for i, (t, has_voice) in enumerate(zip(times_pitch, pitch_values > 0)):
        if not has_voice and i < len(times_pitch)-1:
            plt.axvspan(t, times_pitch[i+1], alpha=0.2, color='red')

    plt.xlabel('Время (секунды)')
    plt.ylabel('Амплитуда')
    plt.title('🌊 ФОРМА ВОЛНЫ (красное - паузы)')
    plt.grid(True, alpha=0.3)

    # 6. ИНФОРМАЦИЯ
    plt.subplot(3, 2, 6)
    plt.axis('off')

    # Оценка качества анализа
    if max_f0 > pitch_ceiling * 0.9:
        quality = "⚠️ Пик близок к верхней границе - возможно удвоение частоты"
    elif max_f0 < pitch_floor * 1.2:
        quality = "⚠️ Пик близок к нижней границе"
    else:
        quality = "✅ Анализ в норме"

    info_text = f"""
    📊 ПОДРОБНАЯ СТАТИСТИКА:

    Минимум: {min_f0:.1f} Гц
    Максимум: {max_f0:.1f} Гц
    Среднее: {mean_f0:.1f} Гц
    Медиана: {median_f0:.1f} Гц
    Стд. отклонение: {std_f0:.1f} Гц
    Размах: {max_f0 - min_f0:.1f} Гц

    Время пика: {max_time:.2f} сек
    Длительность: {clean_times[-1]:.2f} сек
    Пауз: {np.sum(pause_indices)} ({np.sum(pause_indices)/len(pause_indices)*100:.1f}%)

    Метод: Filtered Autocorrelation
    Диапазон: {pitch_floor}-{pitch_ceiling} Hz
    {quality}
    """
    plt.text(0.1, 0.5, info_text, fontsize=12, verticalalignment='center',
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

    plt.tight_layout()
    plt.show()

    # --- АНАЛИЗ ИНТОНАЦИИ ---
    print("\n📊 РЕЗУЛЬТАТЫ АНАЛИЗА (Filtered Autocorrelation)")
    print("-"*60)
    print(f"📈 Максимум: {max_f0:.1f} Гц")
    print(f"📉 Минимум: {min_f0:.1f} Гц")
    print(f"📊 Среднее: {mean_f0:.1f} Гц")
    print(f"📐 Медиана: {median_f0:.1f} Гц")
    print(f"📏 Стд. отклонение: {std_f0:.1f} Гц")
    print(f"⏱️  Пик на {max_time:.2f} сек")
    print(f"⏸️  Пауз: {np.sum(pause_indices)} ({np.sum(pause_indices)/len(pause_indices)*100:.1f}%)")

    # Определяем пол
    if mean_f0 > 165:
        gender = "женский"
    else:
        gender = "мужской"

    print(f"🗣️  Тип голоса: {gender} (по средней частоте {mean_f0:.1f} Гц)")

    # Проверка на возможные ошибки анализа
    if max_f0 > pitch_ceiling * 0.9:
        print(f"\n⚠️ ВНИМАНИЕ: Пик ({max_f0:.1f} Гц) близок к верхней границе ({pitch_ceiling} Гц)")
        print("   Возможно, анализ захватил гармонику вместо основного тона.")
        print(f"   Попробуйте уменьшить Pitch ceiling до {int(pitch_ceiling * 0.7)} Гц")

    # Анализ интонации
    print("\n💡 АНАЛИЗ ИНТОНАЦИИ:")

    # Проверяем, где находится пик
    if max_time < clean_times[-1] * 0.3:
        peak_position = "в начале"
    elif max_time > clean_times[-1] * 0.7:
        peak_position = "в конце"
    else:
        peak_position = "в середине"

    # Оцениваем вариативность
    if std_f0 > 50:
        variability = "высокая (эмоциональная речь)"
    elif std_f0 > 25:
        variability = "средняя"
    else:
        variability = "низкая (монотонная речь)"

    print(f"   📍 Пик интонации: {peak_position} фразы")
    print(f"   📊 Вариативность: {variability}")

    # Определяем тип интонации
    if peak_position == "в конце" and (max_f0 - mean_f0) > 30:
        print("   🔴 ВОПРОСИТЕЛЬНАЯ ИНТОНАЦИЯ (восходящий тон)")
        print(f"   📐 Подъем: {max_f0 - mean_f0:.1f} Гц")
    elif peak_position == "в начале" and (max_f0 - mean_f0) > 40:
        print("   🟡 ЭМОЦИОНАЛЬНОЕ ВОСКЛИЦАНИЕ (резкий подъем в начале)")
    elif std_f0 < 20:
        print("   🔵 МОНОТОННАЯ РЕЧЬ (ровная интонация)")
    else:
        print("   🔵 ПОВЕСТВОВАТЕЛЬНАЯ (нисходящая или ровная)")

    # Даем послушать
    print("\n👂 Прослушать запись:")
    display(Audio(filename))

# Регистрируем функции для вызова из JavaScript
output.register_callback('process_audio', process_recorded)
output.register_callback('select_file', select_file)

# Показываем интерфейс
display(HTML(mode_html))
```
